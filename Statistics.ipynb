````python
# 3)  Write a Python class representing a discrete random variable with methods to calculate its expected
# value and variance.

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initialize the discrete random variable with values and their corresponding probabilities.
        :param values: List of possible values of the random variable.
        :param probabilities: List of corresponding probabilities for each value.
        """
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not all(0 <= p <= 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if not abs(sum(probabilities) - 1.0) < 1e-6:
            raise ValueError("Probabilities must sum to 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the random variable.
        :return: Expected value.
        """
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the random variable.
        :return: Variance.
        """
        mean = self.expected_value()
        return sum((value - mean) ** 2 * prob for value, prob in zip(self.values, self.probabilities))

# Example usage
values = [0, 1, 2, 3, 4]  # Possible values of the random variable
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]  # Corresponding probabilities

rv = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
````

````python
# 4 Create a Python function to generate random samples 6rom a given probabilits distribution (e.g.,
# binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_samples_and_calculate_stats(distribution, params, sample_size=1000):
    """
    Generate random samples from a given distribution and calculate their mean and variance.
    :param distribution: The name of the distribution ('binomial' or 'poisson').
    :param params: A dictionary of parameters required for the distribution.
    :param sample_size: The number of samples to generate (default is 1000).
    :return: A tuple containing the mean and variance of the samples.
    """
    if distribution == 'binomial':
        n = params.get('n')
        p = params.get('p')
        if n is None or p is None:
            raise ValueError("For binomial distribution, 'n' and 'p' parameters are required.")
        samples = np.random.binomial(n, p, sample_size)
    elif distribution == 'poisson':
        lam = params.get('lam')
        if lam is None:
            raise ValueError("For poisson distribution, 'lam' parameter is required.")
        samples = np.random.poisson(lam, sample_size)
    else:
        raise ValueError("Unsupported distribution. Supported distributions are 'binomial' and 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)
    return mean, variance

# Example usage
# Binomial distribution with n=10 and p=0.5
binomial_params = {'n': 10, 'p': 0.5}
mean_binomial, variance_binomial = generate_samples_and_calculate_stats('binomial', binomial_params)
print("Binomial Distribution - Mean:", mean_binomial, "Variance:", variance_binomial)

# Poisson distribution with λ=3
poisson_params = {'lam': 3}
mean_poisson, variance_poisson = generate_samples_and_calculate_stats('poisson', poisson_params)
print("Poisson Distribution - Mean:", mean_poisson, "Variance:", variance_poisson)

import numpy as np
````

````python
# Question 8: Write a Python function to calculate the Probability Density Function (PDF) of a continuous random variable
# for a given normal distribution.

import scipy.stats as stats


def calculate_pdf(value, mean, std_dev):
    """
    Calculate the Probability Density Function (PDF) of a continuous random variable
    for a given normal distribution.

    Parameters:
    value (float): The point at which to evaluate the PDF.
    mean (float): The mean of the normal distribution.
    std_dev (float): The standard deviation of the normal distribution.

    Returns:
    float: The PDF value at the given point.
    """
    # Calculate the PDF using scipy.stats.norm
    pdf_value = stats.norm.pdf(value, loc=mean, scale=std_dev)
    return pdf_value

# Example usage
mean = 0       # Mean of the normal distribution
std_dev = 1    # Standard deviation of the normal distribution
value = 1.5    # Point at which to evaluate the PDF

pdf_value = calculate_pdf(value, mean, std_dev)
print(f"PDF at x = {value}: {pdf_value}")
````

````python
# Question 9: Create a program to calculate the Cumulative Distribution Function (CDF) of an exponential distribution.
import scipy.stats as stats



def calculate_cdf(value, rate):
    """
    Calculate the Cumulative Distribution Function (CDF) of an exponential distribution.

    Parameters:
    value (float): The point at which to evaluate the CDF.
    rate (float): The rate (λ) of the exponential distribution, which is the reciprocal of the mean.

    Returns:
    float: The CDF value at the given point.
    """
    # Calculate the CDF using scipy.stats.expon
    cdf_value = stats.expon.cdf(value, scale=1/rate)
    return cdf_value

# Example usage
rate = 1.0   # Rate (λ) of the exponential distribution
value = 2.0   # Point at which to evaluate the CDF

cdf_value = calculate_cdf(value, rate)
print(f"CDF at x = {value}: {cdf_value}")
````

````python
# Question 10: Write a Python function to calculate the Probability Mass Function (PMF) of a Poisson distribution.

import scipy.stats as stats



def calculate_pmf(k, rate):
    """
    Calculate the Probability Mass Function (PMF) of a Poisson distribution.

    Parameters:
    k (int): The point at which to evaluate the PMF (k must be a non-negative integer).
    rate (float): The rate (λ) of the Poisson distribution, which is the average number of occurrences in a fixed interval.

    Returns:
    float: The PMF value at the given point.
    """
    # Calculate the PMF using scipy.stats.poisson
    pmf_value = stats.poisson.pmf(k, mu=rate)
    return pmf_value

# Example usage
rate = 3.0  # Rate (λ) of the Poisson distribution
k = 4       # Point at which to evaluate the PMF

pmf_value = calculate_pmf(k, rate)
print(f"PMF at k = {k}: {pmf_value}")
````

````python
# Question 11: A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). 
# They collect data from the old and new layouts to compare. 
# Perform a z-test to determine which layout is more successful.
import numpy as np
from statsmodels.stats.proportion import proportions_ztest



# Data: old and new layouts
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate the number of successes and trials for each layout
successes = np.array([old_layout.sum(), new_layout.sum()])
trials = np.array([len(old_layout), len(new_layout)])

# Perform the z-test for proportions
z_stat, p_value = proportions_ztest(successes, trials)

# Output the results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of results
alpha = 0.05  # significance level
if p_value < alpha:
    print("The new layout is significantly different from the old layout.")
else:
    print("There is no significant difference between the new and old layouts.")
````

````python
# Question12: A tutoring service claims that its program improves students' exam scores. 
# A sample of students who participated in the program was taken, and their scores before and after the program were recorded. 
# Use a z-test to determine if the claims made by the tutor are true or false.
import numpy as np
from scipy import stats



# Data: Scores before and after the program
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate differences between scores before and after the program
differences = after_program - before_program

# Compute mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_dev_diff = np.std(differences, ddof=1)
n = len(differences)

# Perform the z-test
z_stat = mean_diff / (std_dev_diff / np.sqrt(n))
p_value = stats.norm.sf(z_stat) * 2  # two-tailed p-value

# Output the results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of results
alpha = 0.05  # significance level
if p_value < alpha:
    print("The tutoring program has a significant effect on improving scores.")
else:
    print("There is no significant effect of the tutoring program on improving scores.")
````

````python
# Question 21
# A political analyst wants to determine if there is a significant association between age groups and voter preferences. Collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences.


import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a contingency table
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2, p_val, dof, expected = chi2_contingency(contingency_table)

# Determine if there's a significant association
alpha = 0.05
significant = p_val < alpha
(chi2, p_val, significant)
````

````python
#  24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The
# company wants to determine if there is a significant difference in customer satisfaction scores among the
# three product versions. They conducted a survey and collected customer satisfaction scores for each
# version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
# difference in customer satisfaction scores.


#   Use the following data:

#   ```python

#   # Sample data: Customer satisfaction scores for each product version

#   standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]

#   premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

#   deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

#   ``
    
import scipy.stats as stats

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(standard_scores, premium_scores, deluxe_scores)

f_statistic, p_value
````
