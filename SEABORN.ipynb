```python
# 1. Scatter Plot to Visualize the Relationship Between Two Variables
 
 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)
x = np.random.rand(100) * 100
y = 0.5 * x + np.random.randn(100) * 10  # Linear relationship with noise

# Create DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['X'], df['Y'], c='blue', marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X vs Y')
plt.grid(True)
plt.show()
```

```python
#2. Visualize the Distribution of a Numerical Variable
 
 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random numbers
np.random.seed(30)
data = np.random.randn(1000)  # Normal distribution

# Create DataFrame
df = pd.DataFrame({'Value': data})

# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Value'], bins=30, kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of a Numerical Variable')
plt.show()
```

```python
#3. Compare Different Categories Based on Numerical Values
 
 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate dataset
np.random.seed(40)
categories = ['Category A', 'Category B', 'Category C']
values = np.random.rand(30) * 100
category_data = np.random.choice(categories, 30)

# Create DataFrame
df = pd.DataFrame({'Category': category_data, 'Value': values})

# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Comparison of Different Categories Based on Numerical Values')
plt.show()
```

```python
#4. Visualize the Distribution of a Numerical Variable Across Different Categories
 
 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate dataset
np.random.seed(50)
categories = ['Category X', 'Category Y', 'Category Z']
values = np.random.rand(150) * 100
category_data = np.random.choice(categories, 150)

# Create DataFrame
df = pd.DataFrame({'Category': category_data, 'Value': values})

# Create violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Value', data=df, palette='muted')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Distribution of a Numerical Variable Across Different Categories')
plt.show()
```

```python
#5. Visualize the Correlation Matrix of a Dataset Using a Heatmap
 
 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset with correlated features
np.random.seed(60)
n = 100
X = np.random.randn(n, 3)
X[:, 1] = X[:, 0] * 0.5 + np.random.randn(n) * 0.1  # Correlated feature
X[:, 2] = X[:, 0] * -0.3 + np.random.randn(n) * 0.2  # Correlated feature

# Create DataFrame
df = pd.DataFrame(X, columns=['Feature1', 'Feature2', 'Feature3'])

# Compute correlation matrix
corr = df.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
