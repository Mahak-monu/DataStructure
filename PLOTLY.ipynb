```python

# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
# np.random.seed(30)
# data = {
#     'X': np.random.uniform(-10, 10, 300),
#     'Y': np.random.uniform(-10, 10, 300),
#     'Z': np.random.uniform(-10, 10, 300)
# }
# df = pd.DataFrame(data)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set the seed for reproducibility
np.random.seed(30)

# Create the dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')

# Label the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Set title
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(20)

# Create the dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to get a matrix format suitable for heatmap
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5)
        
# Label the axes and set the title
plt.xlabel('Month')
plt.ylabel('Day')
plt.title('Sales Variation Across Different Months and Days')

# Show the plot
plt.show()
```

```python
# 4: Generate a 3D surface plot to visualize the function 


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$')
plt.show()
```

```python
# 5: Create a bubble chart to represent each country's population (y-axis), GDP (x-axis), and bubble size proportional to the population

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5)

for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=12, ha='right')

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Population vs GDP')
plt.show()
```
